# P20. Buscar la distancia mínima que existe entre dos locales de McDonald ‘s en el estado de LA (que no comparten dirección). ¿Cuál es la distancia?¿A qué ciudad/es pertenecen? ¿Cuáles son las direcciones de los locales?

```python
mcdonadls_in_la = business[(business['state'] == 'LA') & (business['name'] == 'McDonald\'s' )]
mcdonadls_in_la.head()
```

```python
len(mcdonadls_in_la)
```

```python
mcdonadls_in_la = mcdonadls_in_la[['business_id', 'latitude', 'longitude', 'address', 'city']]
```

```python
mcdonadls_in_la.head()
```

```python

def haversine_np(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees)

    All args must be of equal length.    

    """
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2

    c = 2 * np.arcsin(np.sqrt(a))
    km = 6367 * c
    return km
```

```python
# position = 0
df = mcdonadls_in_la.merge(mcdonadls_in_la, how='cross')
df['distance'] = haversine_np(df['longitude_x'], df['latitude_x'],
                                                           df['longitude_y'], df['latitude_y'])
df.drop(df[df['distance'] == 0].index, inplace=True)
# print(df['distance'].min())
# print(df['distance'].idxmin())
df.loc[[df['distance'].idxmin()]]
```

```python
df.sample(10)
```

## Respuesta
La distancia mas corta es 1.15 km y las direcciones son: 501 Westbank Expy y 1212 Stumpf Blvd, ambos ubicados en la ciudad de Gretna.
