```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset/weatherAUS.csv')
```

```python
sns.set(rc={'figure.figsize':(10,10)}, font_scale=1.5)

weather = df.pivot_table(values='Rainfall', index='WindDir9am', columns='WindDir3pm', aggfunc='sum')

ax = sns.heatmap(weather)

# title
# ax.set_xticklabels(['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
#                   rotation=90)
ax.set(title='Como afecta el cambio de la dirección del viento a la lluvia', xlabel='Dirección del viento a las 3 PM', ylabel='Dirección del viento a las 9 AM')
```
