```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset/weatherAUS.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.sample(10)
```

```python
# sns.set(rc={'figure.figsize':(10,10)}, font_scale=1)
sns.set(rc={"figure.dpi": 150, 'savefig.dpi': 150})

weather = df.pivot_table(values='Rainfall', index='WindDir9am', columns='WindDir3pm', aggfunc='mean')

ax = sns.heatmap(weather, cmap="BuPu", cbar_kws={'label': 'lluvia en mm'})


ax.set(title='Las lluvias se concentran cuando hay viento del Sur Suroeste y del Oeste durante todo el día', xlabel='Dirección del viento a las 3 PM',
       ylabel='Dirección del viento a las 9 AM')
```

```python
df["HumidityMean"] = (df["Humidity9am"] + df["Humidity3pm"]) / 2
```

```python
df = df[df['Date'].notna()]
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['month'] = df['Date'].dt.month
```

```python


# my_palette = {"G": "lightgreen", "PG": "darkgreen", "PG-13":"orange", "R": 'red', "Not Rated": 'gray'}

# ax = sns.barplot(x = 'WindDir9am',
#             y = 'Rainfall',
# #             hue = 'mpaa_rating',
#             data = df)


# ax = sns.violinplot(y=df["RainToday"], x=df["Humidity9am"])
ax = sns.violinplot(x="month", y="HumidityMean", hue="RainToday", data=df, palette="Pastel1", aspect=.1)
ax.set(title='Durante Mayo, Junio y Julio la humedad promedio \n no varía tanto los días de lluvia')
ax.set_xticklabels(['Enero','Febrero','Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'])
ax.set(ylabel='Humedad promedio', xlabel='')

sns.set(rc={"figure.dpi": 200, 'savefig.dpi': 200})
# sns.set(rc={'figure.figsize':(1,1)})
plt.xticks(rotation = 45)

# plt.legend(loc='lower right', title='Llovio')

plt.show()
```

```python
mean_rainnfall_by_location = df.groupby('Location').agg({'Rainfall': 'mean'})
```

```python
len(df['Location'].unique())
```

```python
mean_rainnfall_by_location.sort_values(by='Rainfall', ascending=False)
```

```python
most_rainy_locations = mean_rainnfall_by_location.sort_values(by='Rainfall', ascending=False).head(10)
```

```python
data = df[df['Location'].isin(most_rainy_locations.reset_index()['Location'])]

ax = sns.violinplot(x="Location", y="HumidityMean", hue="RainToday", data=data, palette="Pastel1", aspect=.1)
ax.set(title='Hay ciudades que necesitan una humedad mayor para que llueva')
ax.set(ylabel='Humedad promedio', xlabel='')

sns.set(rc={"figure.dpi": 200, 'savefig.dpi': 200})
# sns.set(rc={'figure.figsize':(1,1)})
plt.xticks(rotation = 45)

# plt.legend(loc='lower right', title='Llovio')

plt.show()
```

```python
less_rainy_locations = mean_rainnfall_by_location.sort_values(by='Rainfall').head(10)
data = df[df['Location'].isin(less_rainy_locations.reset_index()['Location'])]

data['RainToday'] = data['RainToday'].replace({0:'No', 1:'Si'})
data = data.rename(columns={'RainToday':'Llovió'})
```

```python
less_rainy_locations = mean_rainnfall_by_location.sort_values(by='Rainfall').head(10)
data = df[df['Location'].isin(less_rainy_locations.reset_index()['Location'])]

data.loc['RainToday'] = data['RainToday'].replace({0:'No', 1:'Si'})
data = data.rename(columns={'RainToday':'Llovió'})

ax = sns.violinplot(x="Location", y="HumidityMean", hue="Llovió", data=data, palette="Pastel1", aspect=.1)
ax.set(title='En Sale y la zona del aeropuerto Melbourne la humedad \n no es un factor determinante para presentar lluvia')
# ax.set_xticklabels(['Enero','Febrero','Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'])
ax.set(ylabel='Humedad promedio', xlabel='')

sns.set(rc={"figure.dpi": 200, 'savefig.dpi': 200})
# sns.set(rc={'figure.figsize':(1,1)})
plt.xticks(rotation = 45)
# plt.legend(ax.legend_.legendHandles, loc='lower right', labels=['Si', 'No'])
# plt.legend(loc='lower right', title='Llovio')

plt.show()
```
